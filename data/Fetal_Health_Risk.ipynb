# **Model Selection**

```python
# Import necessary libraries
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, recall_score, precision_score, f1_score
from sklearn import metrics
from sklearn.metrics import roc_curve, auc, roc_auc_score
```

```python
# Assign values to features as X and target as y
X=data.drop(["fetal_health"],axis=1)
y=data["fetal_health"]
```

```python
# Display the range of data for each features
plt.figure(figsize=(10,5))
sns.boxenplot(data = data)
plt.xticks(rotation=90)
plt.show()
```

```python
# Data scaling
col_names = list(X.columns)
s_scaler = preprocessing.StandardScaler()
X_df= s_scaler.fit_transform(X)
X_df = pd.DataFrame(X_df, columns=col_names)   
X_df.describe()
```

```python
X_df
```

```python
# Display the range of data after scaling
plt.figure(figsize=(10,5))
sns.boxenplot(data = X_df)
plt.xticks(rotation=90)
plt.show()
```

```python
# Split test and training sets
X_train, X_test, y_train,y_test = train_test_split(X_df,y,test_size=0.3,random_state=42) 

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Buildpipelines of model for various classifiers
pipeline_lr = Pipeline([('lr_classifier', LogisticRegression(max_iter=10000))])
pipeline_dt = Pipeline([('dt_classifier',DecisionTreeClassifier())])
pipeline_rf = Pipeline([('rf_classifier',RandomForestClassifier())])
pipeline_svc = Pipeline([('svc_classifier',SVC())])

# List of all the pipelines
pipelines = [pipeline_lr, pipeline_dt, pipeline_rf, pipeline_svc]

# Dictionary of pipelines and classifier types for ease of reference
pipe_dict = {0: 'Logistic Regression', 1: 'Decision Tree', 2:'RandomForest', 3: 'SVC'}

# Fit the pipelines
for pipe in pipelines:
    pipe.fit(X_train, y_train)

cv_results_accuracy = []
for i, model in enumerate(pipelines):
    cv_score = cross_val_score(model, X_test,y_test, cv=12)
    cv_results_accuracy.append(cv_score)
    print("%s: %f " % (pipe_dict[i], cv_score.mean()))
```

After evaluating the performance of multiple models on the dataset, it shows that random forest model perform the best among the four models with the highest cross validation score.


```python
# Accuracy on test set
pred_rfc = pipeline_rf.predict(X_test)
accuracy = accuracy_score(y_test, pred_rfc)
print("Accuracy on test set:", accuracy)
```

```python
RF_model = RandomForestClassifier()
RF_model.fit(X_train, y_train)

# Test the model on test set
predictions = RF_model.predict(X_test)

score_rf_train = RF_model.score(X_train, y_train)
score_rf_test = RF_model.score(X_test, y_test)

# Predict on the training and test data
train_predictions = RF_model.predict(X_train)
test_predictions = RF_model.predict(X_test)

# Evaluate the accuracy of the model
train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

print("r^2 on training set:", score_rf_train)
print("r^2 on testing set:", score_rf_test)
```

```python
# Classification report for training set
print("Training Set Metrics:")
print(classification_report(y_train, train_predictions))

# Classification report for test set
print("Test Set Metrics:")
print(classification_report(y_test, test_predictions))
```

From the r^2 and the evaluation metrics above, it seems that the model is overfitting the training data. Thus, we apply regularization technique to overcome this.

```python
# Define the parameters for the regularized Random Forest model
rf_params = {
    'n_estimators': 100,
    'max_depth': 7,
    'min_samples_split': 15
}

# Create an instance of the regularized Random Forest model
rf_model = RandomForestClassifier(**rf_params)

# Fit the model to the training data
rf_model.fit(X_train, y_train)

# Predict on the training and test data
train_predictions = rf_model.predict(X_train)
test_predictions = rf_model.predict(X_test)

# Evaluate the accuracy of the model
train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)

print("r^2 on training set:", train_accuracy)
print("r^2 on test set:", test_accuracy)
```

```python
# Classification report for training set
print("Training Set Metrics:")
print(classification_report(y_train, train_predictions))

# Classification report for test set
print("Test Set Metrics:")
print(classification_report(y_test, test_predictions))
```

```python
importances = rf_model.feature_importances_

# Sort the feature importances in descending order
sorted_indices = importances.argsort()[::-1]

# Print feature importances
for i in sorted_indices:
    print(f"Feature: {X.columns[i]}, Importance: {importances[i]}")


# Get the feature names
feature_names = X.columns

# Plot the feature importances
plt.figure(figsize=(10, 6))
plt.bar(range(len(importances)), importances[sorted_indices], tick_label=feature_names[sorted_indices])
plt.xlabel('Features')
plt.ylabel('Importance')
plt.title('Feature Importance')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
